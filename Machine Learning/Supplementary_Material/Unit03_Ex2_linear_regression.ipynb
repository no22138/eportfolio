```python
import matplotlib.pyplot as plt
from scipy import stats

#Create the arrays that represent the values of the x and y axis
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

#Execute a method that returns some important key values of Linear Regression
slope, intercept, r, p, std_err = stats.linregress(x, y)
print("slope", slope)
print("intercept", intercept)
print("r", r)
print("p", p)
print("std_err", std_err)
# measure the correlation 
corr, _ = stats.pearsonr(x, y)
print('Pearsons correlation: %.3f' % corr)

#Create a function that uses the slope and intercept values to return a new value. 
#This new value represents where on the y-axis the corresponding x value will be placed
def myfunc(x):
  return slope * x + intercept

#Run each value of the x array through the function. This will result in a new array with new values for the y-axis
mymodel = list(map(myfunc, x))

#Draw the original scatter plot & the line of linear regression
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

Interpretation : 
- Slope = -1 : for every one unit increase in x we will have approximately 1.75 unit decrease in y (in average).
- intercept : the value of the dependant variable y where the independant variable x is 0 ( when x =0 ==> y=100.3).
- r : pearson coeff ( perfect negative correlation )
- p : p-value : significance level = 0.002 < 0.05 ==> strongly reject the null hypothesis , means that there should be linear regression between x and y
- std_err: 0.45 the standard error value of approximately 0.45 means that the estimated slope of -1.7513 could vary by approximately  + or - 0.45 units on average if we were to repeat the experiment multiple times with different samples of data (same population).

## Predict Future Values

```python
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return intercept + slope * x 

speed = myfunc(10)

print(speed)
```

If x=10 then predicted y is 85.59
